```python
# TO CHANGE FROM false TO False in JSON file

def ReplaceInPlace(filePath, actualWord, correctdWord):
    
    with open(filePath, "r") as file:
        content = file.read()
        
    updated_content = content.replace(actualWord, correctdWord)

    with open(filePath, 'w') as file:
        file.write(updated_content)

    print("Updated {0} to {1} in movieData.py".format(actualWord, correctdWord))
```

```python
import requests

api_key = "48dd50145846e5e031880367400b5eeb"
base_url = "https://api.themoviedb.org/3/discover/movie"
certifications = ['G', 'PG-13', 'PG', 'R', 'Not Rated']
certification_country = 'US'  # Assuming we are interested in US ratings

movies = []

for certification in certifications:
    params = {
        "api_key": api_key,
        "primary_release_date.gte": "2020-01-01",
        "primary_release_date.lte": "2023-12-31",
        "certification_country": certification_country,
        "certification": certification
    }

    response = requests.get(base_url, params=params)

    if response.status_code == 200:
        data = response.json()
        for movie in data.get('results', []):
            title = movie.get('title')
            rating = movie.get('vote_average')
            movie_certification = certification  # Adding certification to the output
            movies.append({
                "Title": title,
                "Rating": rating,
                "Certification": movie_certification
            })
    else:
        print(f"Failed to retrieve data for certification {certification}: {response.status_code}")

# Print the collected movies
for movie in movies:
    print(f"Title: {movie['Title']}, Rating: {movie['Rating']}, Certification: {movie['Certification']}")

```
