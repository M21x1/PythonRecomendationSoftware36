```python
# TO CHANGE FROM false TO False in JSON file

def ReplaceInPlace(filePath, actualWord, correctdWord):
    
    with open(filePath, "r") as file:
        content = file.read()
        
    updated_content = content.replace(actualWord, correctdWord)

    with open(filePath, 'w') as file:
        file.write(updated_content)

    print("Updated {0} to {1} in movieData.py".format(actualWord, correctdWord))
```

```python
filePath = "movieData.py"

actualWord = 'false'
correctdWord = 'False'

ReplaceInPlace(filePath, actualWord, correctdWord)
```
